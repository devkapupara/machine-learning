# Indexing data

```python
A = np.array(np.arange(16)).reshape(4, 4)
A
```

```python
A[1, 2]    # 2nd row, 3rd column = 6
```

```python
A[[1,3]]    # Access only 2nd and 4th rows. Pass in a list of rows you want to access
```

```python
A[:, [0, 2]]    # For all rows, access only the first and 3rd columns
```

Let's try to extract a submatrix of 2nd and 4th row; 1st and 3rd column

```python
A[[1, 3], [0, 2]]
```

What happened? Why did we get [4, 14]? When passed in like that, numpy thinks it is asking for A[1, 0] and A[3, 2] which equals to 4 and 14 respectively.

Use `np.ix_([1, 3], [0, 2])` to create a locator index for the submatrix and then use it as an index.

```python
idx = np.ix_([1, 3], [0, 2])
A[idx]
```

Inefficient but another way of creating it. First get only first and third rows using `A[[1, 3]]` and then use another index notation to grab all rows but only 1st and 3rd columns using `[:, [0, 2]]`

```python
A[[1, 3]][:, [0, 2]]
```

```python
A[1:4:2, 0:3:2]
```

Another way of achieveing the same thing but using splices. `1:4:2` generates `[1, 3]` and `0:3:2` generates `[0:2]`.

# Indexing Practice Problems

Create a data frame consisting of the `weight` and `origin` of the subset of cars with `year` greater than 80.

```python
idx_80 = auto_reindex['year'] > 80
idx_80
```

```python
auto_reindex.loc[idx_80, ['weight', 'origin']]
```

But we can do better and use `lambda` functions.

```python
auto_reindex.loc[lambda df: df['year'] > 80, ['weight', 'origin']]
```

Now get all cars after 1980 and can reach more than 30 miles per gallon

```python
auto_reindex.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30), ['weight', 'origin']]
#auto_reindex.loc[lambda df: (df['year'] > 80) and (df['mpg'] > 30), ['weight', 'origin']]
```

Note that do not use `and` here in the filter query for rows. We need to use the element-wise `&` operator. Same thing when we want any one of the condition to be true, use `|` and not `or`. Element-wise operator `&` and `|` here ensure that the condition(s) are met by all columns of the row.

Now retrieve all `Ford` and `Datsun` cars with `displacement` less than 300.

```python
auto_reindex.loc[lambda df: (df['displacement'] < 300) & (df.index.str.contains('ford') | df.index.str.contains('datsun')), 
                 ['weight', 'origin']]
```

# Zip function

Very useful when we need to iterate over two lists with reference to both objects at same index at the same time. Think of how `enumerate(list)` provides you access to index `i` and the value at `i`, but here, it allows you to traverse two arrays at the same giving you access to ith value from both arrays

```python
total = 0
for value, weight in zip([2, 3, 19], [0.2, 0.3, 0.5]):
    total += value * weight
total
```

# String formatting

For example we may want to loop over the columns of a data frame and print the percent missing in each column. Let’s create a data frame `D` with columns in which 20% of the entries are missing i.e. set to np.nan. We’ll create the values in `D` from a normal distribution with mean 0 and variance 1 using `rng.standard_normal()` and then overwrite some random entries using `rng.choice()`

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))    # (127, 5) matrix with standard normal distribution
A
```

```python
# Create a matrix of same size as A. Assign each cell a value from [0, np.nan]
# with probability of 0.8 for 0 label and 0.2 for np.nan
M = rng.choice([0, np.nan], p=[0.8, 0.2], size=A.shape)
M
```

```python
5 + np.nan
```

Adding `np.nan` to anything results in `np.nan`. So using the fact, we will add `A` and `M` so 20% of the entries will be set to `np.nan` as we wanted to.

```python
A += M
A
```

```python
D = pd.DataFrame(A, columns=['food', 'bar', 'pickle', 'snack', 'popcorn'])
D.head()
```

```python
for col in D.columns:
    print(f'Column {col} has {np.isnan(D[col]).mean():.2%} missing values')
```
